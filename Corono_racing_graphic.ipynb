```python
#Seçilen ülkelerdeki günlük vaka sayısı değişimini çiz
df_countries=df[['Turkey','Italy','Germany','China','US']]
df_countries.plot()
```

```python
#Seçilen ülkelerdeki toplam vaka sayısı
df_countries.sum(axis=0)
```

```python
#Toplam vaka sayısı değişimini grafikleme
df_countries.sum(axis=0).plot(kind='bar')
```

```python
sum_daily=df_countries.cumsum(axis=0) # aynı sütündaki rakamları alt alta toplama
sum_daily
```

```python
sum_daily['Countries_Total']=sum_daily.sum(axis=1)
sum_daily
```

```python
sum_daily.plot()
```

```python
df_4=df[['date','Turkey','Italy','Germany','China','US']]
df_4.info()
```

```python
df_4
```

```python
#Tarihsel olarak sıralı olduğu için date' i index olarak ayarlayabiliriz
df_4['date']=pd.to_datetime(df_4['date'])
#df_4['month']=pd.DatetimeIndex(df_4['date']).month
#df['month']=df['date'].dt.month
df_4.set_index('date',inplace=True) #indexi date olarak ayarla - tarihe göre indexle
df_4
```

```python
#Vaka Sayılarını Kümülatif Toplam Yapma
df_4.cumsum(axis=0).plot()
```

## Yarışan Grafik

```python
df_yarisan=df_4.cumsum(axis=0) # Günlük vakaların kümülatif toplamlarından bir veri seti oluşturma
df_yarisan
```

```python
sns.lineplot(data=df_yarisan)
```
